# Import Library

```python
import numpy as np # Mengimpor pustaka NumPy untuk manipulasi array numerik
import pandas as pd # Mengimpor pustaka Pandas untuk analisis data menggunakan struktur data DataFrame
import matplotlib.pyplot as plt # Mengimpor pustaka Matplotlib untuk visualisasi data
```

```python
# Import modul yang dibutuhkan
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Pemanggilan Data

```python
data = pd.read_csv ('R03_excercises.csv') #membaca data
data #pemanggilan data
```

# Preprocessing Data

```python
X = data['Daily Exercise Time (X)'].values.reshape(-1, 1) # Memilih kolom 'Daily Exercise Time (X)' sebagai fitur (X)
Y = data['Weight Loss (Y)'].values # Memilih kolom 'Weight Loss (Y)' sebagai target (Y)
```

```python
# Misalnya, X dan Y adalah matriks fitur dan vektor target Anda
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Membuat objek model

```python
# Membuat objek model
model = LinearRegression()
```

```python
# Melatih model dengan data pelatihan
model.fit(X_train, Y_train)
```

```python
# Melakukan prediksi dengan data uji
Y_pred = model.predict(X_test)
```

```python
# Menghitung akurasi dengan metrik Mean Squared Error (MSE)
mse = mean_squared_error(Y_test, Y_pred)
print(f'Mean Squared Error (MSE): {mse}')
```

```python
# Menghitung koefisien determinasi (R-squared)
r2 = r2_score(Y_test, Y_pred)
print(f'R-squared (R2): {r2}')
```

# Visualisasi model regresi

```python
# Visualisasi model regresi
plt.scatter(X_test, Y_test, color='yellow', label='Data Asli')
plt.plot(X_test, Y_pred, color='grey', linewidth=2, label='Model Regresi')
plt.xlabel('Daily Exercise Time (X)')
plt.ylabel('Weight Loss (Y)')
plt.legend()
plt.show()
```

```python
# Pelatihan Data
plt.scatter(X_train, Y_train, color='yellow', label='Data Asli')
plt.scatter(X_test, Y_test, color='grey', label='Model Regresi')
plt.xlabel('Daily Exercise Time (X)')
plt.ylabel('Weight Loss (Y)')
plt.legend()
plt.show()
```

```python
# Visualisasi model regresi dengan garis titik-titik penghubung
plt.scatter(X_test, Y_test, color='yellow', label='Data Asli')
plt.plot(X_test, Y_pred, 'o-', color='grey', linewidth=3, label='Model Regresi')
plt.xlabel('Daily Exercise Time (X)')
plt.ylabel('Weight Loss (Y)')
plt.legend()
plt.show()
```
